## Combining Boolean Filtering

```python
over_60 = df["age"] > 60
deck = df["deck"].notnull()
```

```python
over_60_deck_exist = over_60 + deck
```

```python
df[over_60_deck_exist]
```

```python
over_60_and_under_5 = (df["age"] > 60) | (df["age"] < 5)
```

```python
df[over_60_and_under_5]
```
